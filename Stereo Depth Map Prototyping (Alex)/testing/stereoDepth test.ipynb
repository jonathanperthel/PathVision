```python
#import and change images to black/white
#close images (k1 and k2)
left = cv2.imread('l_k1.tif')
left = cv2.cvtColor(left, cv2.COLOR_BGR2GRAY)
right = cv2.imread('r_k1.tif')
right = cv2.cvtColor(right, cv2.COLOR_BGR2GRAY)

#far images
if (False):
    left = cv2.imread('l_f1.tif')
    left = cv2.cvtColor(left, cv2.COLOR_BGR2GRAY)
    right = cv2.imread('r_f1.tif')
    right = cv2.cvtColor(right, cv2.COLOR_BGR2GRAY)

print(left.shape)
print(right.shape)
```

```python
plt.imshow(left, 'gray')
```

```python
plt.imshow(right, 'gray')
```

```python
#perform stereo camera calculation
stereo = cv2.StereoBM_create(numDisparities=16, blockSize=9)
disparity = stereo.compute(left, right)
print(disparity.shape)
plt.imshow(disparity)
```

numDisparities explanation: the disparity search range. For each pixel algorithm will find the best disparity from 0 (default minimum disparity) to numDisparities. The search range can then be shifted by changing the minimum disparity. Parameter mus be positive and divisible by 16.

blockSize explanation:
the linear size of the blocks compared by the algorithm.
The size should be odd (as the block is centered at the current pixel).
Larger block size implies smoother, though less accurate disparity map.
Smaller block size gives more detailed disparity map,
but there is higher chance for algorithm to find a wrong correspondence.

```python
#blockSize seems to be the more important parameter, higher block size means less noise in image,
#this can potentially lead to lower accuracy
for i in range(1,50):
    try:
        if i % 2 != 0:
            #perform stereo camera calculation
            stereo = cv2.StereoBM_create(numDisparities=16, blockSize=i)
            disparity = stereo.compute(left, right)
            print(i)
            plt.imshow(disparity, 'gray')
            plt.show()
    except:
        continue
```

```python
#numDisparities seems to decrease the size of the particles, also higher value causes a bar of blank space
#to take up space moving from left to right in the output window
for i in range(10,300):
    try:
        #perform stereo camera calculation
        stereo = cv2.StereoBM_create(numDisparities=i, blockSize=15)
        disparity = stereo.compute(left, right)
        print(i)
        plt.imshow(disparity, 'gray')
        plt.show()
    except:
        continue
```
